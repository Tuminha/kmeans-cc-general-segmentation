```python
# === TODO ===
# 2) Download the dataset into data/raw using Kaggle CLI
# Hints: !kaggle datasets download -d arjunbhasin2013/ccdata -p data/raw -f "CC GENERAL.csv" --force
# Acceptance: file exists at data/raw/CC GENERAL.csv and is > 1MB

import kagglehub
from pathlib import Path

# Download the dataset using kagglehub
path = kagglehub.dataset_download("arjunbhasin2013/ccdata")

print(f"Dataset downloaded to: {path}")

# Create data/raw directory if it doesn't exist
raw_dir = Path("../data/raw")
raw_dir.mkdir(parents=True, exist_ok=True)

# Copy the CSV file to our data/raw directory
import shutil
source_file = Path(path) / "CC GENERAL.csv"
dest_file = raw_dir / "CC GENERAL.csv"

if source_file.exists():
    shutil.copy2(source_file, dest_file)
    print(f"File copied to: {dest_file}")
    print(f"File size: {dest_file.stat().st_size / (1024*1024):.2f} MB")
else:
    print(f"Warning: Source file not found at {source_file}")
```
